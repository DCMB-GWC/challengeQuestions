## Matplotlib Challenge Problem

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

#### 1. Create a pie chart like the one below of the number of bachelor's degrees awarded in different fields in 2014

```python
majors = [ "Agricultural Sciences", "Biological sciences", "Computer sciences", 
          "Earth sciences","Mathematics and statistics", 
          "Physical sciences", "Psychology", "Social sciences","Engineering"]
number_of_degrees = [28796,109520,56130,22226,6730,22222,118405,177936,93950]
```

```python
plot = plt.pie(number_of_degrees,labels=majors)
plt.title("Bachelors Degrees awarded in 2014")
```

#### 2. Make a bar chart of the number of women and men who got bachelor's degrees in each of these fields in 2014. 

Hint 1: Use ?plt.bar if you are stuck. 

Hint 2: If you're still stuck, try Googling what you're trying to figure out.

Hint 3: This might help you out: https://matplotlib.org/examples/pylab_examples/barchart_demo.html



```python
male = [13271,44811,45986,4134,12715,13407,27554,80813,75324]
female = [15525,64709,10144,2596,9511,8815,90851,97123,18626]
```

```python
fig, ax = plt.subplots()
plt.bar(np.arange(len(male))-0.2,male,width=0.2,color='b',align='center',label='male')
plt.bar(np.arange(len(female)),female,width=0.2,color='g',align='center',label='female')
plt.title("Number of Women and Men awarded Bachelor's degrees (2014)")
plt.ylabel("Number")
ax.set_xticks(np.arange(len(female))-0.5)
ax.set_xticklabels(majors,rotation=70)
plt.show()
```
